# Directed Graphs(Digraphs)


```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 5

# Use seed for reproducibility

G = nx.DiGraph()

# Add nodes

G.add_nodes_from([0,1,2,3,4,5,6,7,8])

# Add self-loops to the remaining nodes

edgelist = [(0, 2),(2,4), (4, 1),(1,3),(3,0),(5,7),(7,5),(6,5),(7,6),(6,9),(4,2),(3,7),(8,4),(5,9),(7,9)]
G.add_edges_from(edgelist)

#pos = nx.spring_layout(G, seed=seed)  # Seed for reproducible layout

#pos = nx.planar_layout(G)

pos = nx.circular_layout(G)
nx.draw(G, pos=pos,with_labels=True)
plt.show()
```

# Bipartite Graphs

```python
import matplotlib.pyplot as plt
import networkx as nx

graph = nx.Graph()

graph.add_nodes_from([9,8,7], bipartite=0) 

# Add the node attribute "bipartite"

graph.add_nodes_from(['red','blue','green'], bipartite=1)

graph.add_edges_from([(9,'red'),(9,'green'),(9,'blue'),(8,'green'),(8,'blue'),(8,'red'),(7,'red'),(7,'blue'),(7,'green')])

top = nx.bipartite.sets(graph)[0]
pos = nx.bipartite_layout(graph, top)
nx.draw(graph, pos=pos, with_labels=True, node_color=['pink','magenta','red','deepskyblue','cyan','green'])
plt.show()
```

Above shows the a bipartite graph, it is a graph where the vertices can be divided into two disjoint sets such that all edges connect a vertex in one set to a vertex in another set.

# Travelling in a Graph

### Walk

```python
import networkx as nx
import matplotlib.pyplot as plt
seed = 6

# Create an empty undirected graph

G = nx.Graph()

# Add nodes to the graph

# G.add_nodes_from([1, 2, 3, 4, 5])

G.add_nodes_from(['BUNZEL','SMED','BCT','LRC','MR','SAFAD','SOCCER'])



edgelist = [('BUNZEL', 'SMED'), ('SMED', 'BCT'), ('BCT','LRC'), ('LRC', 'MR'), ('SOCCER', 'LRC'), ('SOCCER', 'BCT'),('LRC','SAFAD'), ('SMED', 'LRC'),('MR','SAFAD'),('MR','SMED'),('BCT','BUNZEL'),('SAFAD','SOCCER')]
G.add_edges_from(edgelist)
pos = nx.spring_layout(G, seed=seed)
walk = [('SMED', 'BCT'),('BCT','LRC'), ('LRC', 'MR'),('MR','SAFAD'), ('SMED','LRC'), ('LRC','SOCCER')]

# Draw the graph

nx.draw(G, pos=pos, with_labels=True)
nx.draw_networkx_edges(G, pos=pos, edgelist=walk, edge_color='red', width=2)

# Display the plot

plt.show()
```

Walk: SAFAD -> MR -> LRC -> SMED -> BCT -> LRC -> SOCCER<br>
We started the walk at vertex SAFAD and ended at vertex SOCCER. This satisfies as a walk because a walk can have repeated edges and vertices unlike a path.
note: red is pathing

### Euler's Circuit


```python
import networkx as nx
import matplotlib.pyplot as plt
seed = 5
# Create an empty undirected graph
G = nx.Graph()

# Add nodes to the graph
G.add_nodes_from(['A','B','C','D','E','F'])

# Add edges to the graph
edgelist = [('A','B'),('B','C'),('C','D'),('D','F'),('F','B'),('B','E'),('E','D'),('D','A')]
G.add_edges_from(edgelist)
pos = nx.circular_layout(G)


# Draw the graph
nx.draw(G, pos=pos, with_labels=True)

# Display the plot
plt.show()
```

The graph above  contains every vertex and every edge. Therefore it satisfies the condition of the euler's circuit. If there exists a Circuit in the connected graph that contains all the edges of the graph, then that circuit is called as an Euler circuit.

### Euler's Trail

If there exists a Trail in the connected graph that contains all the edges of the graph, then that trail is called as an Euler trail. This graph satisfies that definition.

### Hamiltonian Circuits

```python
import networkx as nx
import matplotlib.pyplot as plt
seed = 6

# Create the graph
G = nx.Graph()
G.add_nodes_from(['BUNZEL','SMED','BCT','LRC','MR','SAFAD','SOCCER'])
edgelist = [('BUNZEL', 'SMED'), ('SMED', 'BCT'), ('BCT','LRC'), ('LRC', 'MR'), ('SOCCER', 'LRC'), ('SOCCER', 'BCT'),('LRC','SAFAD'), ('SMED', 'LRC'),('MR','SAFAD'),('MR','SMED'),('BCT','BUNZEL'),('SAFAD','SOCCER')]
G.add_edges_from(edgelist)
pos = nx.spring_layout(G, seed = seed)

# Find a Hamiltonian circuit
hamiltonian_circuit = [('BUNZEL', 'SMED'),('SMED','MR'), ('MR','SAFAD'), ('SAFAD','LRC'),('LRC','SOCCER'),('SOCCER','BCT'),('BCT','BUNZEL')]

# Draw the graph
nx.draw(G, pos=pos, with_labels=True)

# Highlight the Hamiltonian circuit
nx.draw_networkx_edges(G, pos=pos, edgelist=hamiltonian_circuit, edge_color='red', width=2)

# Display the plot
plt.show()
```

The graph and route above shows an example of a Hamiltonian Circuit. A Hamiltonian circuit is a circuit that visits every vertex once with no repeats. Being a circuit, it must start and end at the same vertex. The route begins with BUNZEL and ends with BUNZEL

# Travelling Salesman Problem

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 3
# Use seed for reproducibility
G = nx.Graph()

# Add nodes
G.add_nodes_from(["A","B","C","D"])
# Add self-loops to the remaining nodes
edgelist = [("A","B"),("A","C"),("A","D"),("B","C"),("B","D"),("C","D"),("A","C"),("B","D")]
G.add_edges_from(edgelist)
pos = nx.planar_layout(G)  # Seed for reproducible layout
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={('A', 'B'): '30', 
                 ('A', 'C'): '50', 
                 ('A', 'D'): '40',
                 ('B', 'C'): '30',
                 ('B', 'D'): '35',
                 ('C', 'D'): '25',},
    font_color='red'
)
nx.draw(G, pos=pos,with_labels=True)
nx.draw(
    G, pos, edge_color='black', width=1, linewidths=1,
    node_size=500, node_color='pink', alpha=0.9,
    with_labels=True)

plt.show()
```

## Solution

```python
import networkx as nx
import matplotlib.pyplot as plt
seed = 7

# Create a graph
G = nx.Graph()
G.add_nodes_from(['BUNZEL','SMED','BCT','LRC','MR','SAFAD','SOCCER'])


edgelist = [('BUNZEL', 'SMED',25),
            ('SMED', 'BCT',20), 
            ('BCT','LRC',40), 
            ('LRC', 'MR',25), 
            ('SOCCER', 'LRC',37), 
            ('SOCCER', 'BCT',24),
            ('LRC','SAFAD',45), 
            ('SMED', 'LRC',23),
            ('MR','SAFAD',15),
            ('MR','SMED',50),
            ('BCT','BUNZEL',32),
            ('SAFAD','SOCCER',16)]
G.add_weighted_edges_from(edgelist)

# Plot the graph
# pos = nx.spectral_layout(G)
pos = nx.spectral_layout(G)
nx.draw(G, pos=pos, with_labels=True)

# Display edge labels
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos=pos, edge_labels=edge_labels, font_color='red', font_size=8)


plt.show()
```

| Route | Total Distance |
| :-: | :-: |
| ${BUNZEL \to SMED \to LRC \to MR \to SAFAD \to SOCCER \to BCT \to BUNZEL}$ | 160 |
| ${BUNZEL \to SMED \to MR \to SAFAD \to SOCCER \to LRC \to BCT \to BUNZEL}$ | 215 |
| ${BUNZEL \to SMED \to MR \to SAFAD \to LRC \to SOCCER \to BCT \to BUNZEL}$ | 228 |
| ${BUNZEL \to SMED \to MR \to LRC \to SAFAD \to SOCCER \to BCT \to BUNZEL}$ | 233 |

- The best route is ${BUNZEL \to SMED \to LRC \to MR \to SAFAD \to SOCCER \to BCT \to BUNZEL}$ with a distance of 160 . The traveling salesman problems finds the shortest Hamiltonian circuit that is possible given a graph. The routes shown above are all possible Hamiltonian circuit and we can see the Hamiltonian circuit with the shortest distance of 160 and longest distance of 233
